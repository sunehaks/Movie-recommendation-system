```python
u = []
pages = []
for i in range(0,len(links)):
    u.append(site+links[i]+'reviews?ref_=tt_urv')
    pages.append(get(u[i]))
len(pages)
```

```python
page_html = []
for i in range(0,len(pages)):
    page_html.append(BeautifulSoup(pages[i].text, 'html.parser'))
```

```python
page_html[:5]
```

```python
tabs1 = []
tabs2 = []
tabs3 = []
tabs4 = []
for i in range(0,len(page_html)):
    tabs1.append(page_html[i].find_all('div', class_ = 'ipl-ratings-bar'))
    tabs2.append(page_html[i].find_all('div', class_ = 'display-name-date'))
    tabs3.append(page_html[i].find_all('div', class_ = 'content'))
    tabs4.append(page_html[i].find_all('div', class_ = 'parent'))
```

```python
#print(len(tabs1[1399]))
'''print(tabs1[0],'\n\n',tabs2[0],'\n\n',tabs3[0])
tabs4'''
len(tabs4)
```

```python
tot_rating = []
tot_user = []
tot_review = []
tot_title = []
for i in range(0,len(tabs1)):
    rating = []
    user = []
    review = []
    title = []
    for j in range(0,len(tabs1[i])):

        rating.append(tabs1[i][j].find('span', class_='').text)
        user.append(tabs2[i][j].find('a').text)
        review.append(tabs3[i][j].find('div',class_ = 'text show-more__control').text)
        title.append(tabs4[i][0].find('a').text)
        #print(title)
        
    tot_rating.append(rating)
    tot_user.append(user)
    tot_review.append(review)
    tot_title.append(title)
```
